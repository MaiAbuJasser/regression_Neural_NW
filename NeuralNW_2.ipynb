```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
#loading Dataset 
from sklearn.datasets import fetch_california_housing
cal_housing = fetch_california_housing()
```

```python
# Explore the dataset
print("Name   of the features: ", cal_housing.feature_names)
print("Shape  of the dataset : ", cal_housing.data.shape)
print("Number of the samples : ", cal_housing.data.shape[0])
print("Number of the features: ", cal_housing.data.shape[1])
print("Number of the targets : ", cal_housing.target.shape)
```

```python
target_col_name = 'PRICE'
df = pd.DataFrame(np.hstack([cal_housing.data, cal_housing.target[:, np.newaxis]]),
                  columns=np.hstack([cal_housing.feature_names, target_col_name]))
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df[cal_housing.feature_names], df[target_col_name], train_size=0.8, random_state=0)
print("Number of the train samples : ", X_train.shape[0])
print("Number of the test samples : ", X_test.shape[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scale =  MinMaxScaler()
X_train = min_max_scale.fit_transform(X_train)
X_test = min_max_scale.transform(X_test)
X_train = pd.DataFrame(data=X_train, index=Y_train.index, columns= cal_housing.feature_names)
X_test = pd.DataFrame(data=X_test, index=Y_test.index, columns= cal_housing.feature_names)
X_train.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop

clf = Sequential()

clf.add(Dense(units=10, activation='relu', input_shape=(8,)))

# for the output layer, you can have one output neuron for regression tasks.
clf.add(Dense(units=1, activation='linear'))  # output layer

# compile classifier
clf.compile(loss='mse', optimizer=RMSprop(lr=0.001), metrics=['mae'])
clf.summary()
```

```python
clf.fit(X_train,Y_train,batch_size = 32,validation_split=0.2, epochs= 1000,verbose = True)
```

```python
#Measuring Model Performance
Y_pred = clf.predict(X_test)
Y_pred.ravel()
```

```python
Y_test.values
```

```python
pd.DataFrame(np.hstack([Y_pred,Y_test.values[:,np.newaxis],np.abs(Y_pred-Y_test.values[:,np.newaxis])]),
      columns = ['Predicted','Actual','MAE'])
```

```python
train_metrics = clf.evaluate(X_train, Y_train)
test_metrics = clf.evaluate(X_test, Y_test)
print(f'Train Loss = {train_metrics[0]}, mae = {train_metrics[1]} ')
print(f'Test Loss = {test_metrics[0]}, mae = {test_metrics[1]} ')
```

# The third tuned model with early stopping

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from keras.callbacks import EarlyStopping

clf = Sequential()

clf.add(Dense(units=10, activation='relu', input_shape=(8,)))

clf.add(Dense(units=7, activation='relu'))


# for the output layer, you can have one output neuron for regression tasks.
clf.add(Dense(units=1, activation='linear'))  # output layer

# compile classifier
clf.compile(loss='mse', optimizer=RMSprop(lr=0.001), metrics=['mae'])
clf.summary()

import time
import numpy as np

# Assuming you have X_train and Y_train datasets

# Define the early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, mode='min', restore_best_weights=True)

start_time = time.time()

# Fit the model with early stopping
history = clf.fit(X_train, Y_train, batch_size=32, validation_split=0.2, epochs=150, verbose=True, callbacks=[early_stopping])

end_time = time.time()
training_time_seconds = end_time - start_time
training_time_minutes = training_time_seconds / 60

print("Training Time: {:.2f} minutes".format(training_time_minutes))

# You can access the best epoch and the corresponding validation loss using the 'history' object.
best_epoch = np.argmin(history.history['val_loss'])
best_val_loss = history.history['val_loss'][best_epoch]

print("Best Validation Loss:", best_val_loss)
```

```python
#Measuring Model Performance
Y_pred = clf.predict(X_test)
pd.DataFrame(np.hstack([Y_pred,Y_test.values[:,np.newaxis],np.abs(Y_pred-Y_test.values[:,np.newaxis])]),
      columns = ['Predicted','Actual','MAE'])
train_metrics = clf.evaluate(X_train, Y_train)
test_metrics = clf.evaluate(X_test, Y_test)
print(f'Train Loss = {train_metrics[0]}, mae = {train_metrics[1]} ')
print(f'Test Loss = {test_metrics[0]}, mae = {test_metrics[1]} ')
```
